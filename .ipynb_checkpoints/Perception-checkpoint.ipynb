# Functional Dependence of Interaction

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from scipy.integrate import odeint
from IPython.display import display, Markdown
from tqdm import tqdm

init_printing()
```

```python
x, y, z, α, β, γ, δ, σ, η, λ, z0, τ = symbols('x y z alpha beta gamma delta sigma eta lambda z_0 tau')
```

```python
xdot = α*y*(1-x) - β*(z0-z)*x*(1-x)
ydot = γ*y*(1-y) - δ*x*y
zdot = τ*(x-z)

fp = solve([xdot, ydot, zdot],(x,y,z))
fp
```

```python
J = Matrix([
    [diff(xdot,x), diff(xdot,y), diff(xdot,z)],
    [diff(ydot,x), diff(ydot,y), diff(ydot,z)],
    [diff(zdot,x), diff(zdot,y), diff(zdot,z)]
])
J
```

```python
J0 = [J.subs([(x,fp[i][0]), (y,fp[i][1]), (z,fp[i][2])]) for i in range(len(fp))]
J0
```

```python
nulcline_x = solve(xdot,y)
nulcline_y = solve(ydot,y)
nulcline_z = solve(zdot,z)
[nulcline_x, nulcline_y, nulcline_z]
```

```python
plt.figure(figsize=(12,12))

plt.axvline(0,color='k')
plt.axhline(0,color='k')

zz = np.linspace(-0.1,1.1,100)
yy1 = np.zeros(len(zz))
yy2 = np.zeros(len(zz))

for ii in range(len(np.linspace(-0.1,1.1,100))):
    yy1[ii] = nulcline_x[0].subs([(x,zz[ii]), (z,zz[ii]), (α,0.1), (z0,0.8), (β,1)])
    yy2[ii] = nulcline_y[1].subs([(x,zz[ii]), (γ,0.035), (δ,0.03)])

plt.plot(zz,yy1,'r')
plt.plot(zz,yy2,'g')
plt.axhline(0, color='g')
plt.axvline(1, color='r')

plt.xlim((-0.1,1.1))
plt.ylim((-0.1,2))

plt.show()
```

```python
def ode_sys(pp, t, α, β, γ, δ, τ, z0):
    
    xx = α*(1-pp[0])*pp[1] - β*(z0-pp[2])*pp[0]*(1-pp[0])
    yy = γ*pp[1]*(1-pp[1]) - δ*pp[0]*pp[1]
    zz = τ*(pp[0]-pp[2])
    
    return [xx,yy,zz]
```

```python
def set_system_parameters(α=0.1, β=1, γ=0.025, δ=0.035, τ=0.01, z0=0.8):
    return { 'α': α, 'β': β, 'γ': γ, 'δ': δ, 'τ': τ, 'z0': z0 }
```

```python
def set_integration_parameters(tf=1000, dt=1.0, x_init=0.9, y_init=0.1, z_init=0.0):
    return { 'tf': tf, 'dt': dt, 'x0': x_init, 'y0': y_init, 'z0': z_init}
```

```python
def numeric_simulation(sys_params, int_params):
    t = np.arange(0, int_params['tf'], int_params['dt'])
    pp = [int_params['x0'], int_params['y0'], int_params['z0']]
    p = (sys_params['α'], sys_params['β'], sys_params['γ'], sys_params['δ'], sys_params['τ'], sys_params['z0'])
    sol = odeint(ode_sys, pp, t, args = p)
    return(t, sol)
```

```python
def compute_basin(xres, yres, total_time=2000, sys_params=None):

    xgrid = np.linspace(0,1,1000)
    ygrid = np.linspace(0,1,1000)

    basin_x = np.zeros((len(xgrid), len(ygrid)))
    basin_y = np.zeros((len(xgrid), len(ygrid)))

    for i in tqdm(range(len(xgrid))):
        for j in range(len(ygrid)):

            if sys_params == None:
                sys_p = set_system_parameters()
            else:
                sys_p = sys_params

            int_p = set_integration_parameters(x0=xgrid[i], y0=ygrid[j], tf=total_time)
            t, sol = numeric_simulation(sys_p, int_p)

            basin_x[i,j] = sol[-1,0]
            basin_y[i,j] = sol[-1,1]
            
    return sys_p['η']*basin_x*basin_y
```

```python
def plot_basin(reward_threshold=0.001):

    low_reward_list = np.where(reward < reward_threshold)
    high_reward_list = np.where(reward > reward_threshold)


    plt.figure(figsize=(12,12))
    plt.plot(low_reward_list[0]/len(xgrid), low_reward_list[1]/len(ygrid), '.r')
    plt.plot(high_reward_list[0]/len(xgrid), high_reward_list[1]/len(ygrid), '.g')

#     xx = np.linspace(-0.1,1.1,100)
#     yy1 = np.zeros(len(xx))
#     yy2 = np.zeros(len(xx))

#     for ii in range(len(np.linspace(-0.1,1.1,100))):
#         yy1[ii] = nulcline_x[0].subs([(x,xx[ii]), (α,0.1), (x0,0.8), (η,1)])
#         yy2[ii] = nulcline_y[1].subs([(x,xx[ii]), (γ,0.035), (δ,0.03)])

#     plt.plot(xx,yy1,'b')
#     plt.plot(xx,yy2,'m')
#     plt.axhline(0, color='m')
#     plt.axvline(1, color='b')

# for i in range(len(z0)):
#     plt.plot(z0[i][0], z0[i][1], '.', color='C'+str(i))

    plt.xlim((0,1))
    plt.ylim((0,1))

    plt.show()
```

```python
plt.figure(figsize=(12,12))
sys_p = set_system_parameters(τ=0.01)
int_p = set_integration_parameters(x_init=0.7, y_init=0.4, z_init=0.8)
t, sol = numeric_simulation(sys_p, int_p)
plt.plot(t,sol[:,0])
plt.plot(t,sol[:,1])
plt.plot(t,sol[:,2])
plt.grid()
plt.show()
```
